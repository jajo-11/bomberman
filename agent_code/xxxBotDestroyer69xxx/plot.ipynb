```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
event_acc = EventAccumulator('runs/Mar27_21-10-33_Lazarus')
event_acc.Reload()
event_acc.Tags()
```

```python
event_acc = EventAccumulator('runs/Mar27_21-10-33_Lazarus')
event_acc.Reload()
eps_0 = event_acc.Scalars("Parameters/Epsilon")
reward_0 = event_acc.Scalars("Stats/Cumulative_Reward")
loss_0 = event_acc.Scalars("Stats/Loss")
score_0 = event_acc.Scalars("Stats/Score")
steps_0 = event_acc.Scalars("Stats/Steps")
```

```python
eps_1 = np.array(list(map(lambda x: x[2], eps_0)))
reward_1 = np.array(list(map(lambda x: x[2], reward_0)))
loss_1 = np.array(list(map(lambda x: x[2], loss_0)))
score_1 = np.array(list(map(lambda x: x[2], score_0)))
steps_1 = np.array(list(map(lambda x: x[2], steps_0)))
```

```python
event_acc = EventAccumulator('runs/Mar26_18-08-11_Lazarus')
event_acc.Reload()
eps_0 = event_acc.Scalars("Parameters/Epsilon")
reward_0 = event_acc.Scalars("Stats/Cumulative_Reward")
loss_0 = event_acc.Scalars("Stats/Loss")
score_0 = event_acc.Scalars("Stats/Score")
steps_0 = event_acc.Scalars("Stats/Steps")
```

```python
eps_0 = np.array(list(map(lambda x: x[2], eps_0)))
reward_0 = np.array(list(map(lambda x: x[2], reward_0)))
loss_0 = np.array(list(map(lambda x: x[2], loss_0)))
score_0 = np.array(list(map(lambda x: x[2], score_0)))
steps_0 = np.array(list(map(lambda x: x[2], steps_0)))
```

```python
event_acc = EventAccumulator('runs/Mar28_17-28-39_Lazarus')
event_acc.Reload()
eps_00 = event_acc.Scalars("Parameters/Epsilon")
reward_00 = event_acc.Scalars("Stats/Cumulative_Reward")
loss_00 = event_acc.Scalars("Stats/Loss")
score_00 = event_acc.Scalars("Stats/Score")
steps_00 = event_acc.Scalars("Stats/Steps")
```

```python
eps_2 = np.array(list(map(lambda x: x[2], eps_00)))
reward_2 = np.array(list(map(lambda x: x[2], reward_00)))
loss_2 = np.array(list(map(lambda x: x[2], loss_00)))
score_2 = np.array(list(map(lambda x: x[2], score_00)))
steps_2 = np.array(list(map(lambda x: x[2], steps_00)))
```

```python
eps = np.concatenate((eps_0, eps_1, eps_2))
reward = np.concatenate((reward_0, reward_1, reward_2))
loss = np.concatenate((loss_0, loss_1, loss_2))
score = np.concatenate((score_0, score_1, score_2))
steps = np.concatenate((steps_0, steps_1, steps_2))
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
plt.figure(figsize=(10,5), dpi=300)
plt.subplot(2, 2, 1)
plt.plot(reward)
plt.plot(moving_average(reward, 1000))
plt.ylim((-100,0))
plt.title("Reward")
plt.subplot(2, 2, 2)
plt.plot(score)
plt.plot(moving_average(score, 1000))
plt.title("Score")
plt.subplot(2, 2, 3)
plt.plot(steps)
plt.plot(moving_average(steps, 1000))
plt.title("Steps")
plt.subplot(2, 2, 4)
plt.plot(loss)
plt.plot(moving_average(loss, 1000))
plt.title("Loss")
plt.tight_layout()
```
